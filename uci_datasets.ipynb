```python
import os
import torch
import tqdm 
import numpy as np 
from pathlib import Path
import pandas as pd

from experiments.data import prepare_dataset, all_datasets

from sgkigp.config import set_seeds 

%load_ext autoreload
%autoreload 2
```

```python
print("Num datasets: ", len(all_datasets))
```

```python
%%time 

def read_data(dataset):
    uci_data_dir = Path(os.path.expanduser("~/sgkigp/data/uci/"))
    data_iter = prepare_dataset(dataset, uci_data_dir=uci_data_dir)
    _, train_x, train_y = next(data_iter)
    _, val_x, val_y = next(data_iter)
    _, test_x, test_y = next(data_iter)

    n_train = train_x.shape[0]
    n_dim = train_x.shape[1]
    n_val = val_x.shape[0]
    n_test = test_x.shape[0]
    
    return [n_dim, n_train, n_val, n_test]

df = pd.DataFrame(columns=('n_dim', 'n_train', 'n_val', 'n_test'))
for dataset in tqdm.tqdm(all_datasets):
    n_dim, n_train, n_val, n_test = read_data(dataset)
    df.loc[dataset] = read_data(dataset)
```
